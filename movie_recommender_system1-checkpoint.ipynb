```python
import ast
```

# Converting the data type from dict into array and fetching only name of genres
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] convert into
#['actions','adventure','sci-fi','fantasy']

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'].iloc[0]
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

# taking the top 3 cast for the model 

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies['crew'].iloc[0]
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L

            
```

# fetching the director name from the crew column

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

# removing the spaces between the name of director

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

# making the overview into list of array

# concatenating the columns into a single coulmns tag

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

# Converts each list of words in 'tags' into a single space-separated string

```python
# Import CountVectorizer for converting text into Bag-of-Words representation
```

```python
# Create CountVectorizer with max 2000 words and removing English stopwords
```

# Function to apply stemming on a text string

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
 new['tags'] = new['tags'].apply(stem)
```

# Function to recommend top 5 similar movies

```python
def recommend(movie):
    movie_idx = new[new['original_title'] == movie].index[0]
    distances = similarity[movie_idx]
    movie_list = sorted(list(enumerate(distances)), reverse=True ,key= lambda x:x[1])[1:6]

    for i in movie_list:
        print(new.iloc[i[0]].original_title)

       
```

```python
recommend('Batman Begins')
```

```python
recommend('Avatar')
```
